# **Ml Project**




# ***Load Data***

```python
import pandas as pd
data = pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv")
data
```

#**`Data Preperation`**

##Data Seperation as X and **Y**

```python
y = data['logS']
print(y)
```

```python
x = data.drop('logS' ,axis=1)
x
```

## ***Data Splitting*** **bold text**

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
x_train
```

```python
x_test
```

# **Model Building**

## **Linear Regression**

### **Training The Model**

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
#training the empty model lr wit x_train and y_train
lr.fit(x_train,y_train)
```

###**Applying The Model To Make Prediction**




```python
y_lr_train_prediction = lr.predict(x_train)
y_lr_test_prediction = lr.predict(x_test)
```

```python
#80% of orig data
print(y_lr_train_prediction)
```

```python
#20% of orig data test data
print(y_lr_test_prediction)
```

**Evaluating Model Performence**

```python
y_train
```

```python
y_lr_train_prediction
```

###**Evaluate Model Performance**





```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train,y_lr_train_prediction)
lr_train_r2 = r2_score(y_train,y_lr_train_prediction)


lr_test_mse = mean_squared_error(y_test,y_lr_test_prediction)
lr_test_r2 = mean_squared_error(y_test,y_lr_test_prediction)
```

```python
print("lr_train_mse: ", lr_train_mse)
print("lr_train_r2: ", lr_train_r2)
print("lr_test_mse: ", lr_test_mse)
print("lr_test_r2: ", lr_test_r2)
```

```python
lr_results = pd.DataFrame(["Linear Regression:" ,lr_train_mse ,lr_train_r2 ,lr_test_mse ,lr_test_r2]).transpose()
```

```python
lr_results.columns = ["Methods","Training Mse","Training R2","Test Mse","Test R2"]
lr_results
```

## **Random Forset**

### **Training The Model**

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train,y_train)
```

### **Applying The Model To Make Prediction**

```python
y_rf_train_prediction = lr.predict(x_train)
y_rf_test_prediction = lr.predict(x_test)
```

```python
#80% of orig data
print(y_rf_train_prediction)
```

```python
#20% of orig data test data
print(y_rf_test_prediction)
```

### **Evaluate Model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train,y_lr_train_prediction)
rf_train_r2 = r2_score(y_train,y_lr_train_prediction)


rf_test_mse = mean_squared_error(y_test,y_lr_test_prediction)
rf_test_r2 = mean_squared_error(y_test,y_lr_test_prediction)
```

```python
print("lr_train_mse: ", rf_train_mse)
print("lr_train_r2: ", rf_train_r2)
print("lr_test_mse: ", rf_test_mse)
print("lr_test_r2: ", rf_test_r2)
```

```python
rf_results = pd.DataFrame(["Random Forest:" ,rf_train_mse ,rf_train_r2 ,rf_test_mse ,rf_test_r2]).transpose()
```

```python
rf_results.columns = ["Methods","Training Mse","Training R2","Test Mse","Test R2"]
rf_results
```

## **Model Comparison**

```python
df_models = pd.concat([lr_results,rf_results],axis=0)
df_models.reset_index(drop=True)
```
